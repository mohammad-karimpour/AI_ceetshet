```python
import pandas as pd

# پانداس برای کار باده داده ها، تجزیه و تجلیل و تمیز کاری آن ها به کار میرود
```

```python
# دو نوع داده وجود دارد

# DataFrame : آرایه جدولی(دو بعدی / سطر و ستون)
df = pd.DataFrame({"name": ['sana','mohammad'], "age": [18,21]})

# Series : آرایه تک بعدی(سطون)
sr = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
```

```python
# متد هاس پانداس
df.head() #=> نمایش 5 سطر اول جدول
df.tail() #=> نمایش 5 سطر آخر جدول
df.dtypes #=> نمایش تایپ هر ستون
df.dtypes.value_counts() #=> نمایش تعداد هر تایپ
df.info() #=> نمایش اطلاعاتی از داده
df.shape #=> نمایش ابعاد دیتا فریم (سطر و ستون)
len(df) #=> نمایش تعداد سطر دیتا فریم 
df['name'].value_counts() #=> نمایش تعداد و تکرار داده های ستون مورد نظر
df['age'].count() #=> نمایش تعداد سطر هایی که NaN نیستند و داده واقعی دارند
df['age'].min(),df['age'].mean(),df['age'].max() #=> نمایش کوچکترین و میانگین و بزرگترین عدد ستون
df['age'].quantile([0.1,0.5]) #=> نمایش آمار متد بالا به صورت دقیق تر و انتخاب درصد خواسی از داده برای استفاده  10% , 50
df.set_index('name') #=> انتخاب یک ستون به عنوان ایدکسیکگ جدول
```

```python
# انتخاب سطر و ستون مشخص

# انتخاب ستون
df.age #=> انتخاب با نام ستون
df['name'] #=> مانند بالا(مناسب برای نام هایی که داخلش فاصله دارند)

# انتخاب سطر
print(df.loc[1])   # بر اساس نام ایندکس
print(df.iloc[0])  # بر اساس شماره ردیف

# انتخاب سطر مورد نظر با فیلتر
print(df[df['age'] > 20])


# انتخاب سطر و ستون مشخص
print(df.loc[1,"name"])   #(ردیف ایندکس 2 با سطونی که اسمش نیم است)
print(df.iloc[:,1])    #(تمام ردیف ها با ستون سن)
```

```python
# خواندن و نوشتن فایل ها

# خواندن فایل CSV
df = pd.read_csv('data.csv') #=> erad_ + csv, html, exeal,...

# نوشتن به فایل CSV
df.to_csv('output.csv', index=False) #=> to_ + csv, html, exeal,...
```

```python
# عملگر های ریاضی ، شرطی
# نیازی به حلقه زدن نیست. پانداس به صورت موازی روی همه سطر ها اعمال میکند

print(df[df['age'] > 20])
print(df['age'] + 120)
```

```python
# تغیر نام با تابع rename
# این تابع نسخه اصلی رو عوض نمی کند

# تغیر نام ستون - title
print(df.rename(columns={'name': 'username'})) #=> key= old value / value=new value

# تغیر نام ایدکس
print(df.rename(index={1: 'Y!!'})) #=> key= old value / value=new value

# به جای دیکشنری میتوان یک تابع داد که ورودی تابع میشه تک تک هدف مورد نظر
def Uper(v):
    return v.upper()

print(df.rename(columns=Uper))
```

```python
# اضافه و حذف کردن ستون

# اضافه کردن
G = [1,4]
df['G'] = G #=> با این روش دیتای اصلی تغیر میکند
df.assign(G=G) #=> با این روش دیتای اصلی تغیر نمی کند

# حذف کردن
df.drop("G", axis="columns") #=> axis مشخص میکنه که از ستون حذف کنه یا از ایندکس که همان سطر است
```

```python
#  انتخاب سطون با فیلتر کردن


# select_dtypes: انتخاب ستون هایی با دیتا تایپ مشخص
# include
df.select_dtypes(include="integer") # فقط ستون های با دیتاتایپ عدد صحیح بده
df.select_dtypes(include=['object','integer']) # انتخاب لیستی از دیتاتایپ ها

# exclude | همه ی دیتاتایپ ها بجز دیتاتایپ ذکر شده
df.select_dtypes(exclude="integer")
df.select_dtypes(exclude=['object','integer'])


# filter: انتخاب ستون بر اساس فیلتر کردن نام ستون

# items: انتخاب ستون های مشخص با نام ستون
df.filter(items=['num','ago'])

# like: ستون هایی که نامشان از این عبارت برخوردار باشد را بم یگرداند
df.filter(like='men')

# regex: انتخاب ستون هایی که با هصورت حساس تر نامشان اتخاب میشود
df.filter(regex='[nN]umbers') #=> numbers, Numbers
```

```python
# پاکسازی دیتا

df.isna() # پیدا کردن داده های خالی
df.fillna(0) # پر کردن داده های خالی
df.dropna() # حذف کردن داده های خراب
# dropna paramers = how=any: just one items is borken of data for delete row, all: must all items of data is broken for delete row
df.drop_duplicates() # حذف داده های تکراری
```
