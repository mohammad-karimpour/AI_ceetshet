```python
import matplotlib.pyplot as plt
# برای نمایش آمار و  نومدار ها به کار می رود

x = [1, 2, 3, 4, 5]
y = [3, 7, 5, 2, 10]

# polt: وظیفه رسم نمودار خطی را دارد
# x:محور افقی   y:محور عمودی
# در صورت دادن یک ورودی به تابع، به صورت خودکار محور افقی را خودش میچیند

plt.plot(x,y)
plt.show()
```

```python
# figure: تعین اندازه (ابعاد) نومدار
plt.figure(figsize=(3,2))

plt.plot(x, y)
plt.show()
```

```python
# شخصی سازی نمودار

plt.title("چونی چاکی")#=> عنوان نمودار

plt.xlabel("روژ")#=> برچسپ محور افقی
plt.ylabel("باران")#=> برچسپ محور عمودی

plt.plot(x, y, color="green", linewidth=3, linestyle="--", marker="s", label="کاربران")#=> color:رنگ خط linewidth:ضخامت خط  linestyle:نوع خط marker:نقاط داده label: راهنمای خط

plt.grid()#=> ایجاد شبکه برای خوانایی بهتر

plt.legend()#=> نمایش راهنمای هر خط(باید داده دارای ویژگی label باشه)
```

```python
# انواع نمودار ها
x = [150, 160, 170, 180, 190]
y = [50, 60, 68, 80, 90]

# Line Plot (نمودار خطی)
plt.plot(x, y)

# Scatter Plot (نمودار پراکندگی) *
# در این نمودار، نقاط به هم وصل نمی‌شوند و هر داده به صورت یک نقطه مستقل نمایش داده می‌شود
plt.scatter(x, y)

# Bar Plot (نمودار ستونی)
plt.bar(x, y)

# Histogram (هیستوگرام) *
plt.hist(y, bins=2) # bins: جزئیات نمایش

# Pie Chart (نمودار دایره‌ای)
labels = ["Cat", "Dog", "Bird"]
sizes = [45,35,20]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
```
